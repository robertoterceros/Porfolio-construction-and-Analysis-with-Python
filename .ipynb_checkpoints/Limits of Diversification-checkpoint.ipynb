## Limits of Diversification

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import edhec_risk_kit as erk
import pandas as pd
import numpy as np

ind_return = erk.get_ind_returns()
ind_nfirms = erk.get_ind_nfirms()
ind_size = erk.get_ind_size()
```

```python
ind_mktcap = ind_nfirms * ind_size
ind_mktcap.shape
```

```python
total_mktcap = ind_mktcap.sum(axis="columns")
```

```python
total_mktcap.plot()
```

```python
ind_capweight = ind_mktcap.divide(total_mktcap, axis="rows")
```

```python
ind_capweight.head()
```

```python
ind_capweight["1926"].sum(axis="columns")
```

```python
ind_capweight[["Fin", "Steel"]].plot()
```

```python
total_market_return = (ind_capweight*ind_return).sum(axis="columns")
```

```python
total_market_return.plot()
```

```python
total_market_index = erk.drawdown(total_market_return).Wealth
```

```python
total_market_index.plot(figsize=(10, 5), title='Total market CapWeighted Index 1926-2018')
```

```python
total_market_index["1980":].plot(figsize=(112, 6))
total_market_index["1980":].rolling(window=36).mean().plot(figsize=(112, 6))
```
